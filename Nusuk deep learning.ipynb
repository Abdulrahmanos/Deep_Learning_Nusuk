```python
import requests
import json

url = "https://haj.gov.sa/assets/api/__%D8%B9%D8%AF%D8%AF%20%D8%A7%D9%84%D8%AA%D8%B5%D8%A7%D8%B1%D9%8A%D8%AD%20%D9%85%D9%86%20%D8%AA%D8%B7%D8%A8%D9%8A%D9%82%20%D9%86%D8%B3%D9%83%201444%D9%87%20%D8%AD%D8%B3%D8%A8%20%D8%A7%D9%84%D9%8A%D9%88%D9%85%20%D9%88%D8%A7%D9%84%D8%AC%D9%86%D8%B3%20%D9%88%D9%86%D9%88%D8%B9%20%D8%A7%D9%84%D8%AA%D8%B5%D8%B1%D9%8A%D8%AD_.json"

response = requests.get(url, verify=False)  # Adding verify=False to bypass SSL certificate verification

if response.status_code == 200:
    try:
        # Manipulate the response text to add square brackets
        modified_data = "[" + response.text + "]"
        data = json.loads(modified_data)

        for record in data:
            permit_count = record["عدد التصاريح من نسك"]
            permit_type = record["نوع التصريح"]
            gender = record["الجنس"]
            date = record["اليوم"]

            print(f"Permit Count: {permit_count}")
            print(f"Permit Type: {permit_type}")
            print(f"Gender: {gender}")
            print(f"Date: {date}")
            print("-----------------------")
    except json.JSONDecodeError as e:
        print("Error parsing JSON:", e)
else:
    print("Failed to retrieve data. Status code:", response.status_code)
```

```python
def convert_hijri_to_gregorian(hijri_date):
    year, month, day = map(int, hijri_date.split('/'))
    gregorian_date = convert.Hijri(year, month, day).to_gregorian()
    return gregorian_date
```

```python
# Define a function to categorize the season
def categorize_season(date):
    if 6 <= date <= 8:
        return 'Summer'
    elif 9 <= date <= 11:
        return 'Fall'
    elif date in [12, 1, 2]:
        return 'Winter'
    else:
        return 'Spring'
    
# Apply the function to create a new 'Season' column
df_english['Season'] = df_english['Month'].apply(categorize_season)
```

```python
from keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Convert the target variable to one-hot encoded format
y_train_categorical = to_categorical(y_train, num_classes=3)
y_val_categorical = to_categorical(y_val, num_classes=3)
y_test_categorical = to_categorical(y_test, num_classes=3)

# Create a classification model
model = Sequential()

# Add input layer
model.add(Dense(units=64, activation='relu', input_dim=26))  # Make sure 'input_dim' is correctly set

# Add one or more hidden layers
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))

# Add output layer (for classification with 3 classes)
model.add(Dense(units=3, activation='softmax'))  # Use 'softmax' activation for multi-class classification

# Compile the model with categorical cross-entropy loss and accuracy metric
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model using the scaled data
model.fit(X_train_scaled, y_train_categorical, epochs=40, batch_size=32, validation_data=(X_val_scaled, y_val_categorical))

# Evaluate the model on the test set using scaled test data
loss, accuracy = model.evaluate(X_test_scaled, y_test_categorical)
print(f"Test Accuracy: {accuracy}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score

y_pred = model.predict(X_test_scaled)

# Convert true labels and predicted labels to class indices
y_true_indices = y_test_categorical.argmax(axis=1)
y_pred_indices = y_pred.argmax(axis=1)

# Calculate accuracy
accuracy = accuracy_score(y_true_indices, y_pred_indices)

# Calculate precision
precision = precision_score(y_true_indices, y_pred_indices, average='weighted')  # You can choose the averaging method you prefer

# Calculate Recall
recall = recall_score(y_true_indices, y_pred_indices, average='weighted')

# calculate F1 score 
f1 = f1_score(y_true_indices, y_pred_indices, average='weighted')

# Print the results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall (Sensitivity):", recall)
print("F1-Score:", f1)
```

```python
from keras.regularizers import l2

# Create a classification model with L2 regularization
model = Sequential()

# Add input layer with L2 regularization
model.add(Dense(units=64, activation='relu', input_dim=26, kernel_regularizer=l2(0.01)))  # Adjust the regularization strength as needed

# Add hidden layers with L2 regularization
model.add(Dense(units=32, activation='relu', kernel_regularizer=l2(0.1)))
model.add(Dense(units=16, activation='relu', kernel_regularizer=l2(0.1)))

# Add output layer (for classification with 3 classes)
model.add(Dense(units=3, activation='softmax'))

# Compile the model with categorical cross-entropy loss and accuracy metric
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the regularized model using the scaled training data
history = model.fit(X_train_scaled, y_train_categorical, epochs=40, batch_size=32, validation_data=(X_val_scaled, y_val_categorical))
```

```python
from tensorflow.keras.regularizers import l2
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import make_scorer, accuracy_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam

# Define your neural network model as a function
def create_model(units=64, activation='relu', learning_rate=0.001, batch_size=32, l2_reg=0.01):
    model = Sequential()
    model.add(Dense(units=units, activation=activation, input_dim=26, kernel_regularizer=l2(l2_reg)))
    model.add(Dense(units=32, activation=activation, kernel_regularizer=l2(l2_reg)))
    model.add(Dense(units=16, activation=activation, kernel_regularizer=l2(l2_reg)))
    model.add(Dense(units=3, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=learning_rate), metrics=['accuracy'])
    return model

# Create the KerasClassifier
model = KerasClassifier(build_fn=create_model, verbose=0)

# Define hyperparameters and their possible values
param_grid = {
    'units': [32, 64, 128],
    'activation': ['relu', 'sigmoid'],
    'learning_rate': [0.001, 0.01, 0.1],
    'batch_size': [16, 32, 64],
    'l2_reg': [0.01, 0.1]  # Regularization strength
}

# Create a scorer for GridSearchCV
scoring = {'accuracy': make_scorer(accuracy_score)}

# Create the GridSearchCV object with refit set to 'accuracy'
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring=scoring, cv=3, verbose=2, refit='accuracy', error_score="raise")

# Fit the grid search to the data
grid_result = grid.fit(X_train_scaled, y_train_multiclass)

# Print the best hyperparameters and corresponding accuracy
print("Best Parameters: ", grid_result.best_params_)
print("Best Accuracy: ", grid_result.best_score_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Evaluate the best-tuned model
best_tuned_model = grid_result.best_estimator_

# Predict the classes on the test set
y_pred_best_tuned = best_tuned_model.predict(X_test_scaled)

# Calculate accuracy
accuracy_best_tuned = accuracy_score(y_true_indices, y_pred_best_tuned)

# Calculate precision
precision_best_tuned = precision_score(y_true_indices, y_pred_best_tuned, average='weighted')

# Calculate Recall
recall_best_tuned = recall_score(y_true_indices, y_pred_best_tuned, average='weighted')

# Calculate F1 score
f1_best_tuned = f1_score(y_true_indices, y_pred_best_tuned, average='weighted')

# Calculate and print the confusion matrix
confusion_best_tuned = confusion_matrix(y_true_indices, y_pred_best_tuned)

# Print the results for the best-tuned model
print("\nBest-Tuned Model Evaluation:")
print("Accuracy:", accuracy_best_tuned)
print("Precision:", precision_best_tuned)
print("Recall (Sensitivity):", recall_best_tuned)
print("F1-Score:", f1_best_tuned)

# Plot the confusion matrix for the best-tuned model
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_best_tuned, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Best-Tuned Model)')
plt.show()
```
